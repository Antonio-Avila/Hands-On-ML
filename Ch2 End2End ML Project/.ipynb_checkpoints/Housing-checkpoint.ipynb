#### Random Sampling

```python
np.random.seed(42)

def split_train_test(data, test_ratio):
    shuffled_ind = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_ind = shuffled_ind[:test_set_size]
    train_ind = shuffled_ind[test_set_size:]
    return data.iloc[train_ind], data.iloc[test_ind]

train_set, test_set = split_train_test(housing, 0.2)
```

```python
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)
```

```python
housing["income_cat"] = pd.cut(housing["median_income"], 
                              bins = [0, 1.5, 3, 4.5, 6, np.inf],
                              labels = [1, 2, 3, 4, 5])

sns.distplot(housing.income_cat)
plt.title("Histogram of Median Income Strata")
plt.xlabel("Income Category")
plt.show()
# Since all strata for median income have different amounts, should stratify sample
#       instead of simple random sample to ensure get an accurate representation of each category
```

#### Stratified Sampling

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)

for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
strat_test_set["income_cat"].value_counts()/len(strat_test_set)

# Looking at the percent error for both stratified and random sampling on the test set
#   Shows stratified does a better job representing the categories 
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis = 1, inplace = True)
    
# Very important to generate a proper test set to assess model. 
```

#### Data Cleaning

Need to fix missing values somehow. 3 options:
- get rid of corresponding districts       ( .dropna() )
- drop entire variable                     ( .drop()   )
- replace with median, mean, zero, or etc  ( .fillna() )


```python
house.dropna(subset = ["total_bedrooms"])  #Option 1

house.drop("total_bedrooms", axis = 1)     #Option 2

median = house["total_bedrooms"].median()  #Option 3
house["total_bedrooms"].fillna(median, inplace = True)
```

If replacing value with meanor median, need to keep that value to replace missing values in entire trainiing set and then use same value for the test set if it also has missing values.

Question: what if there are missing values in the test set but not in the train set? Do you resample so that does not occur?

Scikit-Learn has function that takes care of mising values: SimpleImputer.
 - first need to create and instance of the simpleimputer specifying want to replace each attribute's missing values with the desired statistic (median, mean, etc)

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

# drop categorical variable ocean_prox and create copy of numerical vars (cant take median of categories)
housing_num = house.drop("ocean_proximity", axis = 1)

#now can fit imputer to train data using .fit() function
imputer.fit(housing_num)

print(imputer.statistics_)
print(housing_num.median().values)
```

The way it was first applied using the .fillna() function is significantly less efficient than using an imputer. The .fillna() only filled the missing values in the total_bedrooms variable with its median since it was the only one that had missing values in the training data. This may not always be the case with new data. There may be missing values in any of the other variables. The SimpleImputer is more flexible this way. It computes the median in this case for all variables and stores them in the statistics_ instance variable. Then it can fill in any missing values for all variables with the stored values when necessary, including the new, live data. 

After "training"/fitting imputer, can replace the missing values by using the .transform() function.

```python
X = imputer.transform(housing_num)   # plain NumPy array, want to convert back into dataframe

housing_tr = pd.DataFrame(X, columns = housing_num.columns, index = housing_num.index)
housing_tr.head()
```

Scikit-Learn Design

- Consistency
    - all objects share consistent interface
    
- Estimators
    - object that estimates parameters based on dataset
    - for example, imputer estimates the median or mean using the fit() method
    
- Transformers
    - objects that transform a dataset using the tranform() method
    - ex: imputer transforms data to fill the missing values using the estimators
     - all transformers have convinient method: fit_transform()
         - equivalent to calling fit() then transform() 

- Predictors
    - some estimators can make predictions given a dataset
    - ex: LinearRegression model is a predictor
    - used via the predict() method
    - also has score() method tha measures quality of predictors given a test set

- Inspection
    - all estimator hyperparam can be found via instance var (such as imputer.stratergy)
    - all estiamtors learned params also found in instance var (imputer.statistics_)

- Nonproliferation of Classes
    - datasets represented as NumPy arrays or SciPy sparse matrices
        - reason why we had to turn the array into a dataframe after transforming using the imputer

- Composition


- Sensible Defaults

### Feature Scaling

Always good idea to apply feature scaling to data, many ML algorithims do not perform well when numerical vars have different scales. (Think of how PCA is affected if vars arent standardized).

2 common methods to getting attributes on same scaleL:
 - _min-max scaling_(or _normalization_) 
      - subtract the min value and divide by difference of max and min.
     -  Simple. Values shifted and rescaled so all end up between 0 and 1 
     - transformer _MinMaxScaler_ does this
         - has *feature_range* parameter if dont want range between 0 and 1
 - _standardization_
     - subtracts mean and divides by standard deviation
     - resutls in having 0 mean and 1 variance. 
     - values arent bounded in a range like in normalization
         - some algorithms need values in a range. For ex, NNs expect input values betweeen 0 and 1
     - less affected by outliers
     - transformer _StandardScaler_
     
**Warning**: important to fit scalers to training data only and use them to transform train and test sets (as well as new data.) 
Important since need to calculate the mean and standard deviation. Need to remain blind to the test set to get a proper performace valuation. 


```python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]

        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

Code above is set only to work for the potential attribute bedrooms_per_room. But it can be extended to work for all attributes. It automates the checking of whether it would help ML algorithms perform better. Saves time.

### Transformation Pipelines

Pipelines are helpful when have several steps in that need to be executed in order. 

Next code shows a small simple pipeline processing the numerical variables

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")),
    ('attrib_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
housing_num_tr
```

Pipeline takes list of name/estimator paurs defining a sequence of steps.All except last must be transformers having the .fit_traansform() method ( last one doesnt necessary have to be a transformer since doesnt need the fit_transform() mehtod in it). Names can be anything as long as theyre unique and dont have double underscores

when calling pipelines fit method, it calls the fit_transform() method sequentially at each transformer, taking in the output from the previous transformer sa the parameter for the next call until reaches last one, then calls just fit() method. 

Pipeline exposes same methods as final estimator in pipe.

Instead of handling the categorical and numerical variables separately, as we have so far, can be done all at once in a single pipeline. ColumnTransformer is used for this.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs)
])

housing_prepared = full_pipeline.fit_transform(house)
```

Need to get list of column names for both the numerical and categorical variables before can construct the ColumnTransformer. 
Requires list of tuples containing name, transformer, and list of names(or indices) of columns that transformer should be applied to. 

Recall, OnehotEncoder returned a sparse matrix while the num_pipeline returned a matrix. When get a mix like these, ColumnTransformer estimates the density of final matrix and returns sparse matrix if density is lower than a threshold value.

**Tip**: instead of using transformer, can specify string "srop" if want columsn dropped, or can specify "passthrough" is want columns left untouched. By default, remaining columns not listed are dropped but can set by remainder hyperparameter to any transformer if want them to be handled differently (such as "passthrough")

### Training and Evaluating on Training Set

```python
#first train linear regression model

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, house_labels);
```

```python
#test on few observations from train set

some_data = house.iloc[:5]
some_labels = house_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions:", lin_reg.predict(some_data_prepared))
print("Labels:", list(some_labels))



# To measure the models RMSE

from sklearn.metrics import mean_squared_error

house_preds = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(house_labels, house_preds)
lin_rmse = np.sqrt(lin_mse)
print("RMSE:", lin_rmse)
```

Since the median house values range between \\$120k and \\$265k, a prediction error of about \\$68k is a huge swing. Example of model underfitting on training data.
Means that either the features dont provide enough information to make good predictions or that the model isnt powerful enough. 

To fix underfitting, main ways are:
- select more powerful model
- feed alg with better features
- reduce constraints on model
